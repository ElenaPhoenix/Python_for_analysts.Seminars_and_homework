### 1. 
Изучите распределение домов от наличия вида на набережную

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv('kc_house_data.csv',encoding='cp1251')
df.head()
```

```python
data_pie = df['waterfront'].value_counts()
data_pie
```

```python
plt.figure()
plt.pie(data_pie.values, labels=data_pie.index,autopct= '%.1f%%')
plt.show()
```

### 2. 
Изучите распределение этажей домов

```python
data_floor = df['floors'].value_counts()
data_floor
```

```python
plt.figure()
plt.hist(df['floors'],bins= 30)
plt.xlabel('Этажи')
plt.ylabel('Дома')
plt.title('"Этажи"')
```

```python
plt.figure()
plt.pie(data_floor.values, labels=data_floor.index,autopct= '%.1f%%')
plt.show()
```

```python
plt.figure()
plt.bar(data_floor.index, data_floor.values)
plt.xlabel('Распределение этажей')
plt.ylabel('Кол-во домов')
plt.title('Этажи')
plt.xticks(rotation = 50);
```

### 3. 
Изучите распределение состояния домов 
* Постройте график
* Сделайте выводы

```python
data_con = df['condition'].value_counts()
data_con
```

```python
plt.figure()
plt.pie(data_con.values, labels=data_con.index,autopct= '%.1f%%')
plt.show()
```

```python
plt.figure()
plt.bar(data_con.index, data_con.values)
plt.xlabel('Состояние дома')
plt.ylabel('Кол-во домов')
plt.title('(не)Кондиция')
plt.xticks(rotation = 50);
```

Выводы. Можно сказать, что бОльшая часть домов находится в нормальном среднем состоянии (на "3"). При этом меньше половины от количества нормальных домов приходится на дома в хорошем состоянии, примерно 1/3 от количества домов в хорошем состоянии приходится на дома в отличном состоянии. Домов в неудовлетворительном состоянии ничтожно мало, количество стремится к 0.
