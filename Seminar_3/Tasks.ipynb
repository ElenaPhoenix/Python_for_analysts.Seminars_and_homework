### 4.
Создать новый признак Cpu_Company, который будет содержать только название фирмы, которая произвела CPU (Intel, AMD и т.д)

```python
df['Cpu_Company'] = 0 # Создаем новый признак (колонку), который изначально пуст
df.head()
```

```python
df[['Cpu_Company', 'Cpu']] # внешние [] относятся к df, внутренние [] создают объект "список"
```

```python
def decompos_str (strcpu): # Создаем функцию, чтобы создать обработчик, который будет обрабатывать каждую ячейку колонки Cpu. Нужно сделать декомпозицию, вычленить str до пробела
    return strcpu.split()[0]
df['Cpu_Company'] = df['Cpu'].apply(decompos_str)
df[['Cpu','Cpu_Company']] 
```

```python
df['Cpu_Company'].value_counts() # Сгруппировали все значения CPU_Company, так можем проверить, допустили ли ошибку в функции
```

### 5.
Создать новый признак Memory_Amount, который будет содержать только количество Gb памяти без указания типа носителя (128, 1024, 256 и т.д.)

```python
df['Memory_Amount'] = 0 # Создаем новый признак (колонку), который изначально пуст
df['Memory'].value_counts() # ознакомимся, в какой состоянии столбец Memory, чтобы понять, какой способо извлечения данных подойдет
```

```python
def decompos_mem(str_mem): 
    # Сначала создаем функцию, которая не будет учитывать терабайты. 
    # Split до пробела: str_mem.split()[0]
    # а потом срезом удаляем  буквенные обозначения памяти (GB, TB) до -2 индекса: str_mem.split()[0][:-2]
    # преобразуем строку в INT, вызываем функцию int(): int(str_mem.split()[0][:-2])
    return int(float((str_mem.split()[0][:-2]))) #добавляем возможность делать из вещественного числа целое. сначала выражение преобразуем в float, потом в int
```

```python
# ячейка тестирования def decompos_mem
decompos_mem('1TB HDD ')
```

```python
# ячейка тестирования def decompos_mem
decompos_mem('256GB SSD +  1.0TB Hybrid  ')
```

```python
# ячейка тестирования def decompos_mem
decompos_mem('1.0TB HDD')
```

```python
df['Memory_Amount'] = df['Memory'].apply(decompos_mem) # подставляем decompos_mem в формирование столбца Memory_Amount
```

```python
df.head()
```

### 6.
Создать новый признак Memory_Type, который будет содержать только тип носителя (HDD, SDD и т.д.)

```python
def decompos_type(str_type): # создаем функцию
    return str_type.split()[1]

df['Memory_Type'] = df['Memory'].apply(decompos_type) # подставляем decompos_type в формирование столбца Memory_Type

decompos_type('256GB SSD +  1.0TB Hybrid  ') # тестируем на строке
```

```python
df['Memory_Type'].value_counts()
```

## Задача 3
Создайте датафрейм с клиентами
laptop_id - это индексы датафрейма с ноутбуками

```python
clients = pd.DataFrame({
    'client_id': [45, 32, 67, 33, 43],
    'laptop_id': [506, 398, 710, 120, 1999]
})
df.head()
```

### 1.
Присоедините к таблице clients данные по ноутбукам через метод join

```python
clients_id = clients.set_index('laptop_id') # Для слияния сопоставление таблиц будет по индексам. В табличке clients устанваливаем индексы, чтобы потом соотнести. Метод set_index(имя_признака_который_будет_индексом)
```

```python
clients_id.join(df) # Через метод join() передаем df
```

```python
df.iloc[506] # слияние произошло по индексу!
```


### 2.
Присоедините к таблице clients данные по ноутбукам через метод merge. Это нужно, чтобы понимать, какие ноутбуки покупались клиентами

```python
merged = clients.merge(df, left_on='laptop_id', right_on=df.index, how='left')
merged
```

## Задача 4
Составьте несколько сводных таблиц  (можно использовать Group By / pivot_table)
### 1.
Найдите среднюю стоимость ноутбуков в зависимости от компании производителя. Отсортируйте от меньшей стоимости к большей

```python

df.pivot_table(index='Company',values='Price_euros', aggfunc=[min,max]) # можно передавать несколько функций
```

### 2.
Найдите минимальную, среднюю и максимальную стоимости ноутбуков в зависимости от производителя процессора

```python
import numpy as np
df.pivot_table(index='Cpu_Company',values='Price_euros', aggfunc=[min,max,np.mean])
```

```python
df.groupby('Cpu_Company').agg({'Price_euros': ['min','max','mean']}) # другой способ с помощью groupby
```
