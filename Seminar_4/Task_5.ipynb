Постройте матрицу корреляций для таблицы

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv('laptops_10.csv',encoding='cp1251')
df.head()
```

```python
df.corr(numeric_only=True) # numeric_only=True - использовать только числовые значения dataframe
```

```python
d = df.corr(numeric_only=True) # Корреляция - взаимосвязь между 2 величинами
d = np.round(d,2) # округляем значения до 2 знаков после запятой
d = np.abs(d) # сделали значения абсолютными величинами (неотрицательными)
d
```

```python
plt.figure()
sns.heatmap(d, annot = True, linewidths=0.5, cmap='coolwarm')
```
