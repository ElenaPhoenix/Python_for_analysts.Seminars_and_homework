```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv('laptops_10.csv',encoding='cp1251')
df.head()
```

### 1. 
Изучите количество памяти с помощью matplotlib
* Постройте график
* Назовите график
* Сделайте именование оси x и оси y
* Сделайте выводы

```python
# Здесь нигде plt не связан с dataframe, нам нужно ЯВНО извлечь нужный признак
plt.figure() # Размер фигуры можно указывать или нет, не обязательно. figsize используем, когда нужен график с определенным соотношением сторон
plt.hist(df['Memory_Amount'],bins= 50) # передаем параметры гистограмме: название колонки, на которой будут основываться вычисления, и bins - вертикальные колонки графика. bins можно не укаывать, по умолчанию = 20
plt.xlabel('Объём памяти')
plt.ylabel('Количество') # по у на гистограмме всегда указывается количество/% от количества
plt.title('Память ноутбуков') # название заголовка
#grid() - сетка, на гистограммах не очень наглядна, не используем тут
```

### 2. 
Изучите стоимость ноутбуков с помощью matplotlib
* Постройте график
* Назовите график
* Сделайте именование оси x и оси y
* Сделайте выводы

```python
plt.figure()
plt.hist(df['Price_euros'], bins=50)
plt.xlabel('Стоимость')
plt.ylabel('Кол-во')
plt.title('Распределение стоимости')
```
