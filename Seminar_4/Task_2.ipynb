### 1. 
Изучите распределение типов носителя
* Постройте график
* Сделайте выводы

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv('laptops_10.csv',encoding='cp1251')
df.head()
```

```python
df['Memory_Type'].describe()
```

```python
data_pie = df['Memory_Type'].value_counts()
data_pie
```

```python
plt.figure()
plt.hist(df['Memory_Type'], bins=50)
plt.xlabel('Тип носителя')
plt.ylabel('Кол-во')
plt.title('носители')
```

```python
plt.figure()
plt.pie(data_pie.values, labels=data_pie.index,autopct= '%.1f%%') # 1f -  1 знак после запятой
plt.show() #убирает текст к графику
```

### 2. 
Изучите распределение компаний производителей
* Постройте график
* Сделайте выводы

```python
data_pie = df['Company'].value_counts() # много мелких производителей, поэтому гистограмма будет информативнее круговой диаграммы
data_pie
```

```python
plt.figure()
plt.pie(data_pie.values, labels=data_pie.index,autopct= '%.1f%%')
plt.show()
```

```python
plt.figure()
plt.hist(df['Company'], bins=50)
plt.xlabel('Компании')
plt.ylabel('Кол-во')
plt.title('Компании')
plt.xticks(rotation = 50); # написание названий под углом 50 градусов. ; - также избавляет от необходимости подписывать график
```

```python
plt.figure() # строим по убыванию
plt.bar(data_pie.index, data_pie.values) # бар и гистограмма строят по разным исходным данным
plt.xlabel('Компании')
plt.ylabel('Кол-во')
plt.title('Компании')
plt.xticks(rotation = 50); 
```
