### 1.
Выберите дом клиенту. Клиент хочет дом с видом на набережную(waterfront), как минимум с тремя ванными и с подвалом.

* Сколько вариантов есть у клиента?
* Выведите все варианты таблицей

```python
import pandas as ps
df = ps.read_csv('kc_house_data.csv',encoding='cp1251')
df[(df['waterfront'] == 1) & (df['bathrooms'] >= 3) & (df['sqft_basement'] != 0)].shape[0]
```

```python
df[(df['waterfront'] == 1) & (df['bathrooms'] >= 3) & (df['sqft_basement'] != 0)]
```

### 2.
Выберите дом клиенту.

Клиент хочет дом - либо с очень красивым видом(параметр 'view' - больше) из окна, либо с видом на набережную(waterfront) - и в очень хорошем состоянии(condition) и год постройки не меньше 1980 года. 
* В какой ценовом диапазоне будут дома?

```python
df[((df['view'] == 4) | (df['waterfront'] == 1)) & (df['condition'] == 5) & (df['yr_built'] >= 1980)]['price'].min()
```

```python
df[((df['view'] == 4) | (df['waterfront'] == 1)) & (df['condition'] == 5) & (df['yr_built'] >= 1980)]['price'].max()
```

### 3. 
Выберите дом клиенту. Клиент хочет дом

* без подвала
* с двумя этажами
* стоимостью до 150000. 
* Какая оценка по состоянию у таких домов в среднем?

```python
df[(df['sqft_basement'] == 0) & (df['floors'] == 2) & (df['price'] < 150000)]!=0
```

```python
df[(df['sqft_basement'] == 0) & (df['floors'] == 2) & (df['price'] < 150000)]['condition'].mean()
```
