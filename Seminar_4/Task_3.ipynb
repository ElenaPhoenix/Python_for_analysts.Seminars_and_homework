```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv('laptops_10.csv',encoding='cp1251')
df.head()
```

```python
# Готовим данные для графика
data = (pd.crosstab(df['Company'], df['Cpu_Company'])).drop(columns= 'Samsung').reset_index()
data
```

```python
# Строим столбчатую многорядную диаграмму

plt.figure(figsize=(8, 6))
n_ticks = np.arange(len(data['Intel'])) # определяем кол-во делений
offset = 0.2 # определяем сдвиг
w = 0.4 # определяем ширину столбцов

plt.bar(n_ticks - offset, data['Intel'], width=w) # добавляем сдвиг к кол-ву делений
plt.bar(n_ticks + offset, data['AMD'], width=w)

plt.title('График')
plt.xlabel('Процессоры')
plt.ylabel('Кол-во')
plt.xticks(n_ticks, data['Company'], rotation = 50);  # добавляем метки делений
```
